**Preprocess**

หลังจากการศึกษาและทดลอง preprocess หลายรูปแบบ ได้เลือก 4 รูปแบบที่น่าสนใจมาให้ดูกัน โดยทุกรูปแบบจะมีวิธีการที่ใช้เหมือนกัน คือ Image Denoising, Normalize และ Smoothing Images

แบบที่ 1

Clahe + Canny Edge Detection + Inverted

```python
def preprocess_image1(image):

    denoised_image = cv2.fastNlMeansDenoising(image)
    auto_contrast = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX)
    smoothed_image = cv2.GaussianBlur(auto_contrast, (5, 5), 0)
    clahe = cv2.createCLAHE(clipLimit=5)
    final_img = clahe.apply(smoothed_image)
    edges = cv2.Canny(final_img, 25, 70)
    inverted_image = 255 - edges

    return inverted_image

image = cv2.imread('/content/Dataset/input_com/train/2/9002116R.png', cv2.IMREAD_GRAYSCALE)

preprocessed_image1 = preprocess_image1(image)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(preprocessed_image1, cmap='gray')
axes[1].set_title('Preprocessed Image')
axes[1].axis('off')

plt.show()
```

แบบที่ 2

SD filtering + Clahe

```python
def preprocess_image2(image):
    denoised_image = cv2.fastNlMeansDenoising(image)
    auto_contrast = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX)
    smoothed_image = cv2.GaussianBlur(auto_contrast, (5, 5), 0)
    image32f = np.float32(auto_contrast)

    mu = cv2.blur(image32f, (3, 3))

    mu2 = cv2.blur(np.square(image32f), (3, 3))

    sigma = np.sqrt(mu2 - np.multiply(mu, mu))

    sigma_normalized = cv2.normalize(sigma, None, 0, 255, cv2.NORM_MINMAX)
    sigma_uint8 = np.uint8(sigma_normalized)

    clahe = cv2.createCLAHE(clipLimit=5)
    final_img = clahe.apply(sigma_uint8)

    return final_img

image = cv2.imread('/content/Dataset/input_com/train/2/9002116R.png', cv2.IMREAD_GRAYSCALE)

preprocessed_image2 = preprocess_image2(image)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(preprocessed_image2, cmap='gray')
axes[1].set_title('Preprocessed Image')
axes[1].axis('off')

plt.show()
```

แบบที่ 3

Clahe + Inverted

```python
def preprocess_image3(image):
    if len(image.shape) != 2 or image.dtype != np.uint8:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = np.uint8(image)
    denoised_image = cv2.fastNlMeansDenoising(image)

    auto_contrast = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX)

    clahe = cv2.createCLAHE(clipLimit=5)
    cl1 = clahe.apply(auto_contrast)
    smoothed_image = cv2.GaussianBlur(cl1, (5, 5), 0)
    inverted_image = 255 - smoothed_image
    return inverted_image

image = cv2.imread('/content/Dataset/input_com/train/2/9002116R.png', cv2.IMREAD_GRAYSCALE)

preprocessed_image3 = preprocess_image3(image)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(preprocessed_image3, cmap='gray')
axes[1].set_title('Preprocessed Image')
axes[1].axis('off')

plt.show()
```

แบบที่ 4 ซึ่งได้ผลลัพธ์ดีที่สุด

Equalization + Inverted

```python
# best
def preprocess_image(image):
    if len(image.shape) != 2 or image.dtype != np.uint8:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = np.uint8(image)
    denoised_image = cv2.fastNlMeansDenoising(image)

    auto_contrast = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX)

    equ = cv2.equalizeHist(auto_contrast)
    smoothed_image = cv2.GaussianBlur(equ, (5, 5), 0)
    inverted_image = 255 - smoothed_image
    return  inverted_image

image = cv2.imread('/content/Dataset/input_com/train/2/9002116R.png', cv2.IMREAD_GRAYSCALE)

preprocessed_image = preprocess_image(image)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(preprocessed_image, cmap='gray')
axes[1].set_title('Preprocessed Image')
axes[1].axis('off')

plt.show()
```

**Apply preprocess and augmentation**

```python
class PreprocessTransform(Transform):
    def encodes(self, img: PILImage):
        img_np = np.array(img)
        preprocessed_img = preprocess_image(img_np)
        return PILImage.create(preprocessed_img)

tfms = [Flip(),Contrast()]
```
